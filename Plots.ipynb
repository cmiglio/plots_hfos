```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import functools
sns.set(style="darkgrid")
%matplotlib inline
```

```python
def isState(state):
    a='nose'
    if state == 1:
        a='interictal'
    elif state ==2:
        a='ictal'
    elif state==3:
        a='sleep'
    return a

def conjunction(*conditions):
    return functools.reduce(np.logical_and, conditions)
```

```python
a=isState(1)


results=pd.read_csv('resultados.csv')

# Preprocesado
results['nState']=results['State'].apply(lambda x: isState(x))

results['cond1']=results['trough']/results['HIFP']
results['cond2']=results['HIFP']/results['LOFP']
results['cond3']=results['trough_high']/results['HIFP']
results['ratioHL']=results['mHPow']/results['mLPow']

#ishfo=(results.cond1<0.25 & results.cond2>1 & results.cond3<0.25)
```

```python
a=results.cond1<0.25  
b=results.cond2>1
c=results.cond3<0.25

## Para aÃ±adir la condicion 3:
#hfos=results[conjunction(a,b,c)]
## Sin condicion 3:
hfos = results[conjunction(a,b)]
```

```python
INTERICTAL.reset_index(drop = True, inplace = True) 
labels=INTERICTAL['Labels'].unique()
labels=np.asmatrix(labels)

chancount=np.matlib.zeros(np.shape(labels))
idx=np.arange(0,np.size(labels))
idx=np.asmatrix(idx)
for i, row in enumerate(INTERICTAL.Channel):
    chancount[0,row-1]=chancount[0,row-1]+1

chancount=chancount[0,0:79]  
idx=idx[0,0:79]
deciles=np.percentile(chancount, np.arange(0, 100, 25))

top10=chancount>=deciles[-1]

top10idx=idx[top10]+1

lista=[]
for i in np.nditer(top10idx):
    lista.append(i)
```

```python
hfosTopDecile=hfos[hfos['Channel'].isin(lista)]
```
